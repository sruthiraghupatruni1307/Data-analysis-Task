# Standardise the dataset by using different scaling methods

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Data-Analysis-Using-Python-MB9/master/Day06_02Dec2020/Advertisement.csv")
df.head()
```

```python
df['radio'].value_counts()
```

```python
df['newspaper'].value_counts()
```

```python
df['TV'].max()
```

```python
df.std()
```

```python
import seaborn as sns
plt.title('before applying scaling')
sns.kdeplot(df['TV'])
sns.kdeplot(df['radio'])
sns.kdeplot(df['newspaper'])
sns.kdeplot(df['sales'])
plt.legend()
```

## Standard Scalar
- mean=0 and std=1
    - z=x-mue/sigma!

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
std_data=s.fit_transform(df) # converts df data into standard scaler distribubution data like mean=0 and std=1
std_data
```

```python
std1=pd.DataFrame(std_data,columns=df.columns)
std1.head()
```

```python
std1.describe()
```

```python
import seaborn as sns
plt.title("after applying standard scaling")
sns.kdeplot(std1["TV"])
sns.kdeplot(std1["radio"])
sns.kdeplot(std1["newspaper"])
sns.kdeplot(std1["sales"])
plt.legend()
plt.show()
```

## Minmax Scalar
- +ve features-range[0,1]
- -ve features-range[-1,1]
xsc=(x-xmin)/(xmin-xmax)

```python
from sklearn.preprocessing import MinMaxScaler
s=MinMaxScaler()
minmax_data=s.fit_transform(df)   # converts df data into standard scaler distribubution data like mean=0 and std=1
minmax_data
```

## Robust Scalar
- outliers
- interquartiler range(25%-70%)
v=(v-median)/(p75-p25)

```python
from sklearn.preprocessing import RobustScaler
R=RobustScaler()
R_data=R.fit_transform(df)   # converts df data into standard scaler distribubution data like mean=0 and std=1
R_data
```

```python
r1=pd.DataFrame(R_data,columns=df.columns)
r1.head()
```
